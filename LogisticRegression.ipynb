Logistic Regression sorts inputs into two categories. If the input gives a result that falls below a certain threshold, then it is sorted as class A. If above, it is sorted as class B.

```python
import numpy as np
np.random.seed(1)
```

Dataset has nine columns: 
- sex (0 = male	 1 = female) - dependent variable
-  age	 
- state 
    - michigan	 
    - nebraska	 
    - oklahoma
- income	
 - politics type 
    - conservative	 
    - moderate	 
    - liberal

we load the test and train datasets in two numpy arrays.

```python
train_file = "./gender_prediction_train_data.txt"
train_xy = np.loadtxt(train_file, usecols = range(0,9), delimiter = ",", comments = "#", dtype = np.float32)
train_x = train_xy[:,1:9]
train_y = train_xy[:,0]

train_file = "./gender_prediction_test_data.txt"
test_xy = np.loadtxt(train_file, usecols = range(0,9), delimiter = ",", comments = "#", dtype = np.float32)
test_x = test_xy[:,1:9]
test_y = test_xy[:,0]
```

```python
def sigmoid(z):
    p = 1.0/(1.0 + np.exp(-z))
    return p
```

```python
def logistic_regression(w, b, x):
    z = 0.0
    for i in range(len(w)):
        z += w[i]*x[i] #sum all mults of the inputs with their respective weights 
    z += b #add the bias
    return sigmoid(z) #run the result through the sigmoid and return it. 
```

We also need a loss function. We are using Mean Squared Error for now, though an alternative in Cross Entropy may be implemented later.

```python
def MSE (w, b, data_x, data_y):
    sum = 0.0 
    for i in range(len(data_x)):
        prediction = logistic_regression(w, b, data_x[i]) #run it for all rows in the dataset
        sum += (data_y[i] - prediction)**2 #(how far was the prediction from the real result)
    mse = sum/len(data_x)
    return mse
```

We will now create our model. All the model is, is a set of weights and a bias. 
There are eight predictors in our dataset: age, 3 states, income and 3 political types. So, there will be 8 weights in our set. 

```python
weights = np.zeros(8)
lo = -0.01
hi = 0.01
for i in range(len(weights)):
    weights[i] = (hi - lo)*np.random.random() +lo
bias = 0 

print("initialized weights: ")
print(weights)
```

Now we have everything we need to train our model. Here's the training function:

```python
learning_rate = 0.005
max_epochs = 100000
indices = np.arange(len(train_x)) #ftiaxe ena pinaka me tous AA twn eggrafwn tou training dataset
for epoch in range(max_epochs):
    np.random.shuffle(indices)
    for i in indices:
        x = train_x[i] 
        y = train_y[i] #o stoxos pou prepei na petuxei stin provlepsi (0.0 i 1.0, oi 2 klaseis)
        p = logistic_regression(weights, bias, x)
 
        #update all weights and the bias:
        for j in range (len(weights)):
            weights[j] += learning_rate*x[j]*(y-p)
        bias += learning_rate*(y-p)
    if epoch % 1000 == 0:
        loss = MSE(weights, bias, train_x, train_y)
        print("epoch = %5d  |  loss = %9.4f " % \
          (epoch, loss))
print("Done ")
        
            
```

We also need a way to measure the accuracy of our predictions.

```python
def measure_accuracy(weights, bias, data_x, data_y):
    n_correct = 0
    n_wrong = 0
    for i in range(len(data_x)):
        x = data_x[i] #input no. i 
        y = int(data_y[i]) #ground truth for input no. i
        p = logistic_regression(weights, bias, x)
        if (y == 0 and p<0.5) or (y == 1 and p >= 0.5):
            n_correct += 1
        else:
            n_wrong =+ 1
    accuracy = (n_correct*1.0)/(n_correct+n_wrong)
    return accuracy
```

```python
accuracy_on_training_data = measure_accuracy(weights, bias, train_x, train_y)
print (accuracy_on_training_data)
accuracy_on_test_data = measure_accuracy(weights, bias, test_x, test_y)
print (accuracy_on_test_data)
```
