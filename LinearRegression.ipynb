linear regression makes a line that kinda generally follows the data's trend  f(x)=wx+b

Root Mean Square Error (RMSE) is our way of finding out how well the model did its job in the end. We add up all the errors from the predictions, square them and divide them by the count of the predictions in order to get the mean square error. Then, we sqrt that to get RMSE.

```python
from math import sqrt
import numpy as np

def measure_RMSE(actual, predicted):
    sum_of_errors = 0.0
    for y, y_hat in zip(actual, predicted):
        squared_prediction_error = (y-y_hat)**2
        sum_of_errors += squared_prediction_error

    mean_squared_error = sum_of_errors / float(len(actual))
    return sqrt(mean_squared_error)
```

Now, let's make the training or fitting function. We need to calculate those two coefficients: slope(w) and intercept(b).

Our training function takes pairs of coordinates from our data to calculate the variance and covariance. the λογος of those two gives us w, and we can use that to calculate b next.

```python
def train(x,y):
    n=len(x) #how long should we do this for? well, until we have gone through every pair!
    covariance = 0
    variance = 0
    x_mean = np.mean(x)
    y_mean = np.mean(y) # thank you numpy for making my life easier

    for i in range(n):
        covariance += (x[i] - x_mean)*(y[i]-y_mean)
        variance += (x[i] - x_mean)**2

    slope = covariance/variance
    intercept = y_mean - slope*x_mean

    return slope, intercept  # our W and b! this process is called fitting 
                             # and is like training the regression with our dataset.
```

the point of linear regression is to estimate the values of shit between known points. so, we need a function that takes half a coordinate ( the x from (x,y)) and finds a suitable y for it, using the slope and intercept we found.

```python
def predict (given_x, w, b):
     predicted_y = given_x*w+b
     return predicted_y
```

```python
x = np.arange(1, 51)
y = x*3+5

# Add some random error to the array
y[np.random.randint(0, len(y), size=10)] += np.random.randint(-5, 5)
```

now, lets try to run it on the toy dataset!

```python
w, b = train(x,y)
y_hat  = predict(x, w, b)
print (w, b)
print ("predicted y (y_hat) for each x in the dataset:")
print (y_hat)

print ("predicted y for a x not in the dataset:")
y_hat_notFound  = predict(3.7, w, b)
print(y_hat_notFound)
```
